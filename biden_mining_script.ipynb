```python
auth = tweepy.OAuthHandler(api_key, api_secret)
auth.set_access_token(access_token, access_token_secret)
api = tweepy.API(auth)
```

```python
'''Twitter mining function to pull ~3200 tweets from a given user  '''
def mine_tweets (api, user_id, oldest_id):
    
#     tweets_init = api.user_timeline(screen_name = user_id, 
#                                      include_rts = False, 
#                                      tweet_mode = 'extended',
#                                      wait_on_rate_limit = True)
    
    all_tweets = []
#     oldest_id = tweets_init[-1].id
#     print(oldest_id)
#     while (len(all_tweets) <= 5000):
    curr = oldest_id 
    while (len(all_tweets) <= 3000):  
        tweets = tweepy.Cursor(api.user_timeline, 
                                    screen_name = user_id,
                                    include_rts = False,
                                    tweet_mode = 'extended',
                                    max_id = oldest_id,
                                    wait_on_rate_limit_notify = True).items()

        print("before for loop")
        for tweet in tweets:

            curr = tweet.id
            if not tweet in all_tweets:
                all_tweets.append(tweet)

    print('Number of tweets downloaded {}'.format(len(all_tweets)))
        
    return all_tweets

```
